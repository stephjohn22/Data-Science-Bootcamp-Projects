```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt',sep=' ')
```

```python
# Calculating the average income for each Ethnicity in the dataset
average_income_by_ethnicity = df.groupby('Ethnicity')['Income'].mean()
print(f"{average_income_by_ethnicity}\n")
print(f"""The ethnicity with the highest average income is African American people with an average income of {df[df.Ethnicity == "African American"].loc[:,"Income"].mean()}.
The ethnicity with the second highest average income is Asian people with an average income of {df[df.Ethnicity == "Asian"].loc[:,"Income"].mean()}.
The ethnicity with the lowest average income is Caucasian people with an average income of {df[df.Ethnicity == "Caucasian"].loc[:,"Income"].mean()}.
""")
```

```python
# Calculating whether Married people or single people have a higher balance
average_balance_by_marital_status = df.groupby('Married')['Balance'].mean()
print(f"{average_balance_by_marital_status}\n")
print(f"On average, single people have a higher balance than married people.\nThe average balance of a single person is {df[df.Married == "No"].loc[:,"Balance"].mean()} whereas the average balance of a married person is {df[df.Married == "Yes"].loc[:,"Balance"].mean()}.")
```

```python
# Selecting the highest income in the dataset
highest_income = df.loc[:,'Income'].max()
print(f"The highest income in the dataset is {highest_income}.")
```

```python
# Selecting the lowest income in the dataset
lowest_income = df.loc[:,"Income"].min()
print(f"The lowest income in the dataset is {lowest_income}.")
```

```python
# Calculating the amount of cards recorded in the dataset
number_of_cards = df.loc[:,"Cards"].sum()
print(f"The number of cards that have been recorded in the dataset is {number_of_cards}.")
```

```python
# Calculating how many females in the dataset vs how many males
gender_num = df.loc[:,'Gender'].value_counts()
print(f"{gender_num}\n")
print(f"There are more females in the dataset than males.\nThere are {df['Gender'].value_counts().loc['Female']} females and {df['Gender'].value_counts().loc['Male']} males.")
```
