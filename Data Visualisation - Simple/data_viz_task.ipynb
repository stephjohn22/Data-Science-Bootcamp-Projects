```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile

# Selecting the neccessary rows from the dataframe
selected_manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
selected_cars = car_data[car_data['Manufacturer'].isin(selected_manufacturers)]

# Extracting data for box plot
x_box = selected_cars['Manufacturer']
y_box = selected_cars['Rev.per.mile']

# Create box plot
sns.boxplot(x = x_box, y = y_box, color = 'r')
plt.title('Revs per Mile for Selected Manufacturers')
plt.ylabel('Revs per Mile')
plt.xlabel('Manufacturer')
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


```python
# Filter the dataset for Audi, Hyundai, Suzuki, and Toyota
selected_manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
selected_cars = car_data[car_data['Manufacturer'].isin(selected_manufacturers)]

# Find the car with the highest revs per mile
car_with_highest_revs = selected_cars[selected_cars['Rev.per.mile'] == selected_cars['Rev.per.mile'].max()]

# Print the manufacturer and car model with the highest revs per mile
print("Manufacturer with the car having the highest revs per mile:")
print(car_with_highest_revs[['Manufacturer', 'Model', 'Rev.per.mile']])
```

**Boxplot answer:**  
Toyota has the car with the highest revs per mile

### 2. Histogram

```python
# Histogram of MPG

# Extracting data for histogram
mpg_city = car_data['MPG.city']
mpg_highway = car_data['MPG.highway']

# Create histogram
sns.histplot(mpg_city, color='b', alpha=0.5, label='MPG in City')
sns.histplot(mpg_highway, color='r', alpha=0.5, label='MPG on Highway')
plt.title('Distribution of MPG in City and on Highway')
plt.xlabel('Miles per Gallon (MPG)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**  
It is generally more fuel efficient to drive in the city than on the highway because less MPG is required in the city.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

# Extracting data for line plot
x_line = car_data['Wheelbase']
y_line = car_data['Turn.circle']

# Create line plot
sns.lineplot(x=x_line, y=y_line)
plt.title('Relationship between Wheelbase and Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
There is a positive correlation between wheelbase and turning circle. As the wheelbase increases, its turning circles also increases. This indicates that the vehicle needs more space to turn if it has a larger wheelbase.

### 4. Barplot

```python
# Barplot for mean horsepower

# Extracting data for bar plot
x_bar = car_data['Type']
y_bar = car_data['Horsepower']

#  Create bar plot
sns.barplot(x=x_bar, y=y_bar, estimator=np.mean)
plt.title('Mean Horsepower for Each Car Type')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**  
Larger cars are more likely to have more horsepower than small and compact cars. However, a car being the largest car size does not neccessarily mean it will have the most horsepower. A van is the biggest vehicle type but a midsize and a large car has more horsepower.
